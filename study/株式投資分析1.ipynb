# 株式評価モデル

テキストの序盤、市場の効率性とそれを否定するアノマリー（理論的根拠がないものの、経験的に観測できるマーケットの規則性）の存在について議論がされる。

アノマリーの例（AI検索）

* カレンダーアノマリー(1月効果: 毎年1月に株価が上昇しやすいとされる現象。特に、小型株のパフォーマンスが高い)
* バリューアノマリー(PBR（株価純資産倍率）やPER（株価収益率）が低い「割安株」が、市場全体よりも高いリターンを示す傾向。)
* ハロウィン効果（Sell in May and Go Away）(5月から10月までのパフォーマンスが低調で、11月から4月までの方が高いとされるアノマリー。)

実際市場が効率的であれば、株価は常に**本源的価値**と一致しているはずである。

つまり、実際市場が効率的であれば株価を見れば企業のすべてがわかってしまい、投資家が超過収益を得る機会は存在しない。

しかし、そんなことはないはずで本源的価値と株価には乖離があるはずである。

そのギャップを見つけるために特に**絶対価値評価モデル**を学び企業の価値を測定することが今回の勉強の目的。

また、絶対価値評価モデルよりも簡便で、実務であったり個別株などを購入する人々が良く利用する**相対価値評価モデル**も学ぶ。




# 絶対価値評価モデル
4つ（2種類×2種類）のモデルを学ぶ。

* 企業価値評価モデル　or　株式価値評価モデル
* 割引キャッシュフローモデル　or　残余利益モデル


### 企業価値評価モデル
評価対象を負債も含んだ企業価値全体とする評価モデル。

企業価値を自然にモデル化することが可能というメリットがある。

デメリットは、株式価値を得るためには、企業価値の推計結果から負債価値を減じる手順が必要で、最終的な興味が株式の投資価値にある場合には手順が複雑で余計な推計誤差も入る。

### 株式価値評価モデル
評価対象を株式部分のみを対象とする評価モデル。

本源的価値を直接推定することができる明快さがある一方、将来の配当やROEなどを推計する際に、企業の配当政策や資本政策による影響を考慮する必要がある。

### 割引キャッシュフロー型
シンプルだが、将来のキャッシュフローの前提をどのように置くかによって結果がぶれるデメリットが存在する。

### 残余利益モデル
企業価値を現時点での資産価値と将来企業が生み出す経済的付加価値の合計として示す。

# DDM（Dividend Discount Model）：配当割引モデル
株式価値評価モデル　×　割引キャッシュフローモデル


| **項目**   | **内容**                                              |
|------------|-------------------------------------------------------|
| **計算式** | $$V = \sum^n_{i=1} \frac{\text{D}_i}{(1 + k)^i}$$            |
| **説明**   | 将来の配当期待値を現在価値に割り引くことで株価を算出。|
| **適用例** | 配当を安定的に支払う成熟企業。                         |

---

# 残余利益モデル
株式価値評価モデル　×　残余利益モデル

| **項目**   | **内容**                                              |
|------------|-------------------------------------------------------|
| **計算式** | $$V= B_0 + \sum_{i=1}^n \frac{E_i - kB_{i-1}}{(1 + k)^i}$$|
| **説明**   | 残余利益（純利益から資本コストを引いたもの）を現在価値に割り引いて株価を算出。                     |
| **適用例** | 配当やキャッシュフローが不安定な企業。                  |

---


# DCF(Discount Cash Flow )：割引キャッシュフローモデル
企業価値評価モデル　×　割引キャッシュフローモデル

| **項目**   | **内容**                                                                                     |
|------------|----------------------------------------------------------------------------------------------|
| **計算式** | $$V_f = \sum_{t=1}^n \frac{\text{CF}_t}{(1 + r)^t}$$ |
| **説明**   | 将来のキャッシュフロー（CF）を割引率で現在価値に割り引き、企業全体の価値を算出。                   |
| **適用例** | 安定したキャッシュフローが見込まれる企業。                                                    |

---

# EVA(Economic Value Added)：経済付加価値モデル

企業価値評価モデル　×　残余利益モデル

| **項目**   | **内容**                                              |
|------------|-------------------------------------------------------|
| **計算式** | $$ V_f = IC_0 + \sum^n_{i=1} \frac{NOPAT_i - k_f IC_{i-1}}{(1+k_f)^i}$$ |
| **説明**   | 税引後営業利益（NOPAT）から投下資本コストを引いた経済的利益を評価。これを現在価値に割り引くことで企業の価値を算出。 |
| **適用例** | 資本効率や収益性を強調したい場合に使用。                 |

## 相対価値評価モデル
代表的なモデル

### PER(株価収益率)
$$
PER = \frac{\text{株価}}{\text{1株当たりの純利益}}
$$

### PBR(株価純粋資産倍率)
$$
PBR = \frac{\text{株価}}{\text{1株当たりの純資産}}
$$

### 配当利回り
$$
\text{配当利回り}=\frac{\text{1株当たり配当}}{\text{株価}}
$$
