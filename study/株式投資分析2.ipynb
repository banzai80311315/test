## 準備：等比数列の公式
等比数列とは、各項が前の項に一定の非零数（公比）を掛けたものとなる数列のことを指す。具体的には、数列 $ a_1, a_2, a_3, \ldots $ が等比数列であるためには、次の条件を満たす必要がある。

$$
a_{n} = a_{n-1} \cdot r \quad (n \geq 2)
$$

そのnまでの和は
$$
S_n = a_1 \frac{1 - r^n}{1 - r} \quad (r \neq 1)
$$

具体例）

初項 $ a_1 = 2 $、公比 $ r = 3 $ の等比数列を考える。この場合、最初の5項は$2, 6, 18, 54, 162$と続く。

この数列の和は

$$
S_5 = 2 \frac{1 - 3^5}{1 - 3} = 2 \frac{1 - 243}{-2} = 242
$$

## おまけ（天才ガウス君）
ある日の授業でガウス君は先生から「1から100の数字を足したら総和はいくらになりますか？」と質問されました。

ガウス君は間髪入れずに「5050」と答えました。

先生は驚きます。ガウス君はどうしてすぐに答えられたのでしょうか？

私であれば電卓でひたすら足し算をしてしまいますが、ガウス君の計算は以下です。

$$
1,2,3,...,98,99,100
$$

という数列に対して、最初と最後を足し続けます。すると、

$1 + 100 = 101$ , $2 + 99 = 101$ , $3+98=101$ , ....

と101が続きます。この場合には101が50個ありますので、$50 × 101 = 5050$と計算できます。

これは高校数学で等比数列と一緒に習う等差数列の総和公式と同じです。

$$
S_n = \frac{n}{2}(a + l)
$$


$aは初項、lは末項、nは数列の個数$、今回の場合は$a=1,l=100,n/2=50$ということでした。

## DDM（Dividend Discount Model）：配当割引モデル

### 債券と株式の違い

| **項目**       | **債券**                     | **株式**                     |
|----------------|-----------------------------|-----------------------------|
| **資金提供の形態** | 借金（返済義務あり）         | 出資（返済義務なし）         |
| **利息/配当**    | 固定の利息を支払う           | 配当は利益に応じて決定       |
| **経営権**       | 投資家に経営権なし           | 株主に経営権あり（議決権を持つ場合が多い） |

### 割引配当モデルの基本

株式価値評価モデル　×　割引キャッシュフローモデル


| **項目**   | **内容**                                              |
|------------|-------------------------------------------------------|
| **計算式** | $$V = \sum^{∞}_{i=1} \frac{\text{D}_i}{(1 + k)^i}$$            |
| **説明**   | 将来の配当期待値を現在価値に割り引くことで株価を算出。|
| **適用例** | 配当を安定的に支払う成熟企業。                         |

---

$D_i$は将来の$i$期先の配当期待値、$k$は割引率。

$k$について、投資家サイドからすれば株式に投資するにあたって要求する投資収益率。

企業サイドからすれば株式で資金調達を行ったことに対して支払わなければならいないコストになる。そのため、割引率$k$を株主資本コストと呼ぶ。



### CAPM割引率の決定

割引率は、投資家がその株式に期待するリターンで**資本資産価格モデル（CAPM）**を用いて計算するというのが過去問ででる。

$$
k = \text{E}[R_p] = \beta \left( \text{E}[R_m] - R_f \right) +  R_f
$$

- $ R_f $: 無リスク金利（例：国債利回り）
- $ \beta $: 株式のベータ値（市場全体に対する感応度）
- $ \text{E}[R_m] $: 市場ポートフォリオの期待リターン
- $ \text{E}[R_m] - R_f$: 市場うポートフォリオのリスクプレミアム

※ベータ
$\beta =\frac{\text{Cov}[R_m,R_p]}{\text{Var}[R_m]}$



## ゼロ成長モデル

$D_i = D_j$という仮定を置いた場合を割引配当モデルにおけるゼロ成長モデルと呼ぶ。

$$
\begin{aligned}
V &= \dfrac{D}{1+k} + \dfrac{D}{(1+k)^2} + \dots \\
  &= \dfrac{D}{1 + k} \left\{1 + \dfrac{1}{1+k} + \dfrac{1}{(1+k)^2} + \dots \right\}
\end{aligned}
$$

これは初項が1、公比が$\dfrac{1}{1+k}$の無限公比数列になるため
$$
V = D\dfrac{1}{ 1 - \frac{1}{1+k}} = \frac{D}{k}
$$

1株当たり50円の配当を永久に払う企業について、投資家がこの企業の株式に対し10%の投資収益率を期待している場合には
$$
V = \frac{50}{0.1} = 500
$$

## 定率成長モデル
$D_i = D_{i-1}(1+g)$という仮定を置く場合を定率成長モデルという。

$D_i = D_1(1+g)^{i-1}$となることを利用して

\begin{aligned}
V &= \sum^{∞}_{i=1} \frac{\text{D}_i}{(1 + k)^i}  \\
  &= \sum^{∞}_{i=1} \frac{D_1(1+g)^{i-1}}{(1 + k)^i}  \\
  &= \dfrac{D_1}{1 + k} \left\{  \sum^{∞}_{i=1} \left(\frac{1+g}{1 + k} \right)^{i-1}\right\}
\end{aligned}

カッコ内は初項が1、公比が$\dfrac{1+g}{1+k}$の無限公比数列になるため
$$
V = \dfrac{D_1}{1+k}\dfrac{1}{ 1 - \frac{1+g}{1+k}} = \frac{D}{k-g}
$$

ぐぐるとでる割引配当モデルはこの式が出てくる
