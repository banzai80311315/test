## フランチャイズ価値モデル
サスティナブル成長率と同様、将来にわたってROEが一定で、外部の資金調達を行わない企業を想定。

定率成長率モデルを以下のように書き換えると
\begin{aligned}
V &= \frac{D}{k-g} \\
  &= \frac{ROE}{k}B + \frac{ROE - k}{k} \frac{g}{k-g}B
\end{aligned}

第一項:フランチャイズファクター

$\frac{ROE}{k}B$は利益を再投資せず全額配当にした場合の企業価値を示す。

第二項:成長等価

$\frac{ROE - k}{k} \frac{g}{k-g}B$は再投資したことにより増加する企業価値を示す。

※投資家が期待しているだけの利益を上げられない場合($ROE - k$ < 0)には、再投資しても価値は上昇しない。




【確認】

$d = 1$となり$D =dE = E$となる。

次に、$g = (1 - d)ROE = 0$となるため、$V = \frac{E}{k} = \frac{ROE}{k}B$



【確認】

\begin{aligned}
V &= \frac{ROE}{k}B + \frac{ROE - k}{k} \frac{g}{k-g}B \\
  &= \frac{B}{k(k - g)}(k ROE - gROE + gROE - kg)\\
  &= \frac{ROE - g}{k - g }B \\
  &= \frac{D}{k-g}
\end{aligned}


# 残余利益モデル（割引超過モデル）
残余利益;$R_1 = E_1- k B_{0} = (ROE - k)B_0$を基に計算される企業価値。付加価値の総和を評価に含めるモデル
$$
V= B_0 + \sum_{i=1}^∞ \frac{E_i - kB_{i-1}}{(1 + k)^i}
$$

サープラス関係$B_i = B_{i-1} + E_i - D_i$が成り立っている状況では、配当割引モデルと同じ値となる
\begin{aligned}
V &= \sum_i \frac{D_i}{(1 + k)^i} \\
  &= \sum_i \frac{B_{i-1} + E_i - B_i}{(1+k)^i}\\
  &= \sum_i \frac{B_{i-1} + +kB_{i-1} - kB_{i-1} +E_i - B_i}{(1+k)^i}B \\
  &= \sum_i \left( \frac{B_{i-1}}{(1 + k)^{i-1}} - \frac{B_i}{(1 + k)^i} \right)+\sum_i \frac{E_i - kB_{i-1}}{(1 + k)^i} \\
  &= B_0 + \sum_{i=1}^∞ \frac{E_i - kB_{i-1}}{(1 + k)^i}
\end{aligned}

# 割引キャッシュフローモデル
負債を含めた企業価値全体を評価するのが割引キャッシュフローモデル。

$$
V_f = \sum_{t=1}^∞ \frac{\text{CF}_t}{(1 + k_f)^t}
$$

良く$CF_t$で使われるものは、以下のフリーキャッシュフロー
$$
フリーキャッシュフロー = 営業利益(1 - 税率) + 減価償却費 - 設備投資額 - 運転資本増加額
$$

さらに資本コストも株主の要求収益率に加えて、債権者の要求収益率も以下のように加重平均として加味する。

$$
k_f = \frac{E}{E+D}k_e + \frac{D}{D + E}(1-\tau)k_d
$$

ここで、$E$は株主資本、$D$を有利子負債、$k_e$を株主資本コスト、$k_d$を負債コスト、$\tau$実行利率（法人税率など）

# EVAモデル
$$
V_f = IC_0 + \sum^n_{i=1} \frac{NOPAT_i - k_f IC_{i-1}}{(1+k_f)^i}
$$

$NOPAT=営業利益×(1−税率)$

$IC(投下資本)= 有利子負債 + 株主資本 + 繰延税金 + 少数株主持分 + 連結調整勘定 + 有利子負債以外の固定負債$

株式の残余利益モデルと同じノリで、総資本コストを上回る税引き後の事業利益をEVA(Economic Value Added)の総和をEVAモデルという。
