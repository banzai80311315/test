# 債券投資分析4
## 利回り曲線
利回り曲線、あるいはイールドカーブ(Yield Curve)とは債券の**残存年数**と**最終利回り**の関係を示す。

特に、

* 割引債の残存年数と最終利回りの関係を示すのが「スポット・レート・カーブ」
* 価格が額面に等しい債券の残存年数と最終利回りの関係を示すのが「パー・レート・カーブ」

などが存在する。

これらの利回り曲線から、金融資産の金利と残存期間の関係やその構造、すなわち「**金利の期間構造**」を理解することが重要である。


## 様々な利回り曲線
### スポット・レート・カーブ
スポット・レート・カーブとは、割引債の利回り曲線のことを指す。

スポットレートカーブは、**将来のキャッシュフローを複利ベースで現在価値に割り引くための割引率を示す曲線**と考えてよい。←
山田さんへの質問：過去問でなぜこれが割引率になるのか？

スポットレートの定義は、額面$F$、残存期間$T$の割引債の価格が$P$であるとすると、期間$T$のスポットレートは
$$
r_T = \left( \frac{F}{P} \right)^{\frac{1}{T}} - 1
$$

また、額面1円換算した割引債の価格のことを「割引関数」または「ディスカウント・ファクター」といい、満期$T$のディスカウント・ファクター$d_T$の定義は
$$
d_T = \frac{1}{(1 + r_T)^T}
$$

ここで一つの見方として、固定利付債は割引債の組み合わせで書けることを記載する。

残存期間$T$年、額面$F$円、クーポン$C$の固定利付債の価格$P$は
$$
P = Cd_1 + Cd_2 + … + (C+F)d_T
$$

の組み合わせとみなせる。


スポット・レート$r_t$が以下のように与えられれば

$r_1 = 1.85％,r_2 = 2.20％,r_3 = 2.48％ , r_4 = 2.75％ , r_5 = 3.03％$

例えば残存期間5年、額面100円、クーポンレート3％の固定利付債の理論価格は

$$
P = \frac{3}{1.0185} + \frac{3}{(1.0220)^2} + … + \frac{103}{(1.0303)^5} = 100.02
$$

### フォワード・レート・カーブ
将来の金利をフォワード・レート・カーブと呼ぶ。

$s$年から$t$年にかけてのフォワード・レート$f(s,t)$は、それぞれの期間に対応したスポット・レート$r_s$、$r_t$を用いて、以下のように定義される

$$
f(s , t) = \left( \frac{ (1 + r_{spot,t})^t}{(1 + r_{spot,s})^s} \right)^{\frac{1}{t-s}}-1
$$

$u$年後スタートのフォワード・レートを使う場合にも同様に

$$
f(s , t) = \left( \frac{ (1 + f_{(u,t)})^t}{(1 + f_{(u,s)})^s} \right)^{\frac{1}{t-s}}-1
$$

### パー・レート・カーブ
パー債券の利回りのことをパー・レート・カーブと呼ぶ。

パー債券は$F=P$という条件があるので評価価格は
$$
F = \sum_{t=1}^T\frac{C}{(1+r_{spot,t})^t} + \frac{F}{(1 + r_{spot,T})^T}
$$

パー・レート$r_{par}$は$r_{par} = \frac{C}{F}$と考えればよいので結果的に
$$
r_{par} = \frac{1 - \frac{1}{(1 + r_{spot,T})^T}}{\left( \sum^T_{t=1} \frac{1}{(1 + r_{spot,t})^t} \right)}
$$

と書くことができる。

### 各レートの位置関係
スポットレートカーブが右肩上がりの時には、フォワード＞スポット＞パー

スポットレートカーブが左肩上がりの時には、フォワード＜スポット＜パー


# 利回り曲線の変動
【用語】

順イールド：利回り曲線が右上がりのとき

逆イールド：利回り曲線が右下がりのとき

プロの投資家は現在のイールドカーブの形状を確認して、今後どのように形状が変化するかを予想し、その要因分析を行い、短期債や長期債を組み合わせながら再建の取引を行う。

### イールドカーブ全体が上や下へシフト

**上にシフト**：債券相場全体の金利上昇を予想できる

**下にシフト**：債券相場全体の金利低下を予想できる

### 特定の年限を中心にイールドカーブにこぶしができる

**上に凸**：残存期間の短い債券の利回りが上昇して、長い債券の利回りが低下（短期債から長期債に資金が流れる）

**下に凸**：残存期間の長い債券の利回りが上昇して、短い債券の利回りが低下（長期債から短期債に資金が流れる）

### 短期金利は変わらずイールドカーブが立つ、寝る

**立つ**：中・長期債に対する需要の低下を予想（金利が上昇し、利回りが上昇）

**寝る**：中・長期債に対する需要の増加を予想（金利が低会し、利回りが低下）
