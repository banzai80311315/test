# 債券投資分析3
## 様々な利回り尺度
以下教科書引用。

債券の利回り、すなわち債券投資収益は次の3つの要素からなる。

1. クーポン収入：利息収入。**インカムゲイン**と呼ぶ。
2. 償還差益：償還価格と販売価格の差。**キャピタルゲインorキャピタルロス**と呼ぶ。
3. クーポンの再投資収入：受け取りクーポンの再投資からえられる収入。

複利ベースの最終利回りは、3.の再投資利回りが最終利回りと一致すると仮定した場合の満期までの1年あたり複利利回りに相当する。

↑この文章はおそらく「複利ベースの最終利回り」を「平均利回り」と理解できますよということ？？

以下からその他代表的な利回り尺度を教科書に沿って説明する。

### 直接利回り
直接利回りは、インカムゲインにのみ着目した利回り尺度である。

一年間で得られるクーポン収入$C$、債券価格$P$とすると

$$
直接利回り = \frac{C}{P}
$$

具体例）
額面100円、クーポンレートが5%の債券を考える。この場合、年間のクーポン支払いは5円。もしこの債券が市場で90円で取引されているとすると、直接利回りは次のように計算できる。

$$
\frac{5}{90}= 0.0555・・・
$$

キャピタルゲインなどを考慮しない部分的な指標であることに注意する。

### 単利最終利回り
直接利回りにキャピタルゲインが含まれていないことを簡便的に補正した利回り尺度。

クーポン$C$、額面$F$、満期までの年数$T$、債券価格$P$として

$$
単利最終利回り = \frac{C + \frac{F - P}{T}}{P}
$$

### 実効利回り
期中に得られるクーポンの再投資利回りに関して、投資家が予想した利回りを用いる。

以下教科書引用

最終利回りは、キャッシュフローの時間的な価値を勘案した内部収益率であるが、実際には期中に得られるクーポンを最終利回りと同じ利率で再投資できるとは限らないため、当該債権を満期まで保有した場合に投資額がいくらに増えるかを必ずしも表すことにはならない。

予想される再投資利回りを$r_E$として
$$
P(1 + r)^T = C(1 + r_E)^{T-1} + C(1 + r_E)^{T-2} + … +C(1 + r_E) + C + F
$$

を満たす$r$を実効利回りと定義する。

計算をすると
$$
r = \sqrt[T]{\frac{C\frac{(1 + r_E)^T - 1}{r_E} + F}{P}}-1
$$

例えば、額面100、クーポンレート3%、満期5年のパー債券($P=F$)の最終利回りは3%であるが、期中のクーポンが2%の利回りでしか再投資できないと考えた場合には

$$
r =  \sqrt[5]{\frac{3\frac{(1 + 0.02)^T - 1}{0.02} + 100}{100}}-1 \simeq 2.94％
$$

以下教科書引用（わかりやすいかも）

長期金利より短期金利の方が低いことが多い現実世界では、クーポンの再投資機関が債権の満期よりも短期であることを考慮すると、最終利回りよりも低い金利での再投資機会を予想した「実効利回り」の方がより現実的な収益性の予想になる。

### 保有期間利回り
購入から売却までの期間の利回りを保有期間利回りと呼ぶ。

債券保有期間を$m$、債券の購入価格を$P$、売却価格を$P_m$、年1回のクーポンを$C$とすると、保有期間利回り$r$は
$$
P = \frac{C}{1 + r} + \frac{C}{(1 + r)^2} + … + \frac{C + P_m}{(1 + r)^m}
$$

## 利回り変化と債券価格
債券投資に伴う代表的なリスクとして
1. 金利変動リスク
2. 再投資リスク
3. 期限前償還リスク
4. 信用リスク

このうち、金利変動リスクによって債券価格がどう変化するのかを詳細に見ていく。

考えたいことは

「金利が$\Delta r$だけ動いたら、債券価格の変化幅$\Delta P$はどの程度になるか」

が、近似をうまくやる方法が細かく書かれているだけの印象なので興味のある人が読むだけにとどめる。

デュレーションは1次近似、コンベキシティは2次近似というだけ。
