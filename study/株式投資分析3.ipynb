## 定率成長モデル
$$
V = \dfrac{D_1}{1+k}\dfrac{1}{ 1 - \frac{1+g}{1+k}} = \frac{D}{k-g}
$$

$k$は収益率（資本コスト）、$g$は成長率、$D$は配当

過去問などでは資本コストにCAPMから求められる期待収益率を利用する。

成長率には一定の仮定をおいて、今回学ぶ**サスティナブル成長率**を利用する。

## CAPMの過去問
(中小企業診断士 H29 問20)

CAPMが成立する市場において、マーケット・ポートフォリオの期待収益率が6％、安全利子率が1 ％のとき、当該資産の期待収益率が10 ％となるベータ値として、最も適切なものはどれか。

$$
k = \text{E}[R_p] = \beta \left( \text{E}[R_m] - R_f \right) +  R_f
$$

より
$$
\beta = \frac{\text{0.1 - 0.01}}{\text{0.06 - 0.01}} = 1.8
$$

## 単語の準備
### EPS(Earnings Per Share):1株当たりの純利益

### BPS(Book value Per Share):1株当たりの純資産
企業の財産の内、返済義務のない資金
### ROE(Return On Equity)
$$
\text{ROE} = \frac{\text{純利益}}{\text{純資産}} = \frac{\text{EPS}}{\text{BPS}}
$$

### PBR（Price-to-Book Ratio、株価純資産倍率）
$$
\text{PBR} = \frac{\text{株価}}{\text{1株当たり純資産}} = \frac{\text{株価}}{\text{BPS}}
$$

## サスティナブル成長率
サスティナブル成長率とは企業が外部から資金調達することなしに達成可能な成長率のこと。

$$
g = ROE \ (1 - d)
$$

$d$は配当性向

## 解説
### 仮定：クリーン・サープラス関係
企業が増資を行わずに、純資産の増加が内部留保のみとする。すると、企業は内部留保だけ成長することから以下の関係式が成り立つ。

期末の純資産$B_1$は、期初の純資産$B_0$ に内部留保(純利益-総配当金)$E_0 - D^{all}_0$だけ加えたものになる。
$$
B_1 - B_0 = (E_0 - D^{all}_0)
$$

つまり、損益計算書の純利益部分について、配当以外の内部留保がそのまま来季の貸借対照表の純資産になることを仮定している。

#### (おまけ)来期の純資産に影響を与える要素

現実の企業では、来期の純資産に影響を与えるものは内部留保だけでなく以下の要素も加味される

- **内部留保**（当期純利益のうち配当されなかった部分）
- **株式発行**による新たな資金調達
- **配当支払い**による利益剰余金の減少
- **評価換算差額**や**未実現利益**の影響
- **減損**や**償却**による資産の減少
- **自己株式の取得**による純資産の減少
- **外貨建て資産の評価差額**（為替影響）



### 純資産成長率
$$
\text{純資産成長率} = \frac{B_1 - B_0}{B_0}
$$

サープラス関係から計算すると
\begin{aligned}
\frac{B_1 - B_0}{B_0} &= \frac{E_0 - D^{all}_0}{B_0}  \\
  &= \frac{E_0}{B_0} \left(1 - \frac{D^{all}_0}{E_0}\right) \\
  &= ROE(1-d) = \text{サスティナブル成長率}
\end{aligned}

#### 過去問
(中小企業診断士 R4 問17)

以下の資料に基づき計算したサステナブル成長率（内部留保のみを事業に投資した場合の純資産の成長率）として、最も適切なものを下記の解答群から選べ。

| **項目**   | **金額（単位：万円）** |
|------------|----------------------|
| **売上**   | 5,000          |
| **当期純利益** | 200          |
| **総資産** | 4,000          |
| **純資産** | 1,000          |
| **配当**   | 80          |

$$
ROE = \frac{E}{B} = 0.2
$$

$$
d = \frac{D}{E} =0.4
$$

$$
g = 0.12
$$


### 定理
サスティナブル成長率が成り立つ状況（ROE一定、配当性向一定、財務レバレッジ一定）の場合には、サスティナブル成長率、純資産成長率、利益成長率、配当成長率が等しい。

（確認）

純資産成長率がサスティナブル成長率と一致
\begin{aligned}
\frac{E_1 - E_0}{E_0} &= \frac{B_1 ROE - E_0}{E_0}  \ \ (\because ROEの定義)\\
  &= \frac{(B_0 + E_0(1 - d))ROE}{E_0} - 1 \ \ (\because クリーンサープラス関係)\\
  &= \frac{B_0}{E_0}ROE -  (1 - d) ROE - 1 \\
  &= 1 - (1 - d) ROE - 1 = (1 - d)ROE \ \ (\because ROE一定)
\end{aligned}

純資産成長率が配当成長率に一致
\begin{aligned}
\frac{D^{all}_1 - D^{all}_0}{D^{all}_0} &= \frac{E_1/d - E_0/d}{E_0/d}  \ \ (\because 配当性向一定)\\
  &= \frac{E_1 - E_0}{E_0}
\end{aligned}

定率成長モデルの成長率がサスティナブル成長率に一致
\begin{aligned}
\frac{D^{all}_1 - D^{all}_0}{D^{all}_0} &= \frac{(1 + g)D_0 - D_0}{D_0}  \ \ (\because 定率成長モデルの定義)\\
  &= 1+ g - 1 = g
\end{aligned}
