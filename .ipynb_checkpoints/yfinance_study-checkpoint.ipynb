## クイックスタートを試してみる

### 基本情報の取得info

```python
# ホンダの株式データを取得
## Tickerクラスのインスタンスを生成
ticker = yf.Ticker("7267.T")
```

```python
# 基本情報の取得
info = ticker.info
```

```python
# そもそもの型を確認⇒辞書型
type(info)
```

Pythonの辞書型（dict）は、キーと値のペアを持つデータ構造です。データの追加、更新、取得が簡単にできます。

```python
import pprint
pprint.pprint(info, width=40)
```

```python
print(f"{info['longName']}")
print(f"時価総額: {info['marketCap']:,}円")  # カンマ区切り
print(f"通貨: {info['currency']}")
print(f"PBR: {info['priceToBook']:.2f}")  # 小数点2桁に整形
## これはpandasで見やすくする関数を作る
```

```python
#これは自分で作成した関数
import format_stock_info as fs
```

```python
df = fs.format_stock_info(info)
type(df)
```

```python
print(df)
```

### 時系列データhistory

```python
# 株価履歴を取得
"""
5d , 1mo , 3mo , 1y etc
"""
df = ticker.history(period="3mo")
type(df) # pandas`s DataFrame
```

```python
"""
Open  : 始値 
High  : 高値
Low   : 低値
Close : 終値
Volume: 取引量
"""
print(df)
```

```python
# ファイナンスデータ可視化用ライブラリ
import mplfinance as mpf
```

```python
"""
緑（陽線）：終値が始値より高い（上昇）
赤（陰線）：終値が始値より低い（下落）
"""
# インデックスを日付に設定（mplfinance 用）
df.index = df.index.tz_localize(None)

# ローソク足チャートを描画
mpf.plot(df, type="candle", style="yahoo", volume=True, ylabel="price (JPY)")
```

```python
mpf.plot(df, type="candle", style="charles", volume=True, mav=(5, 25), ylabel="price (JPY)")
```
